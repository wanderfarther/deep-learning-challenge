## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df = application_df.drop(columns=['EIN', 'NAME'])
```

```python
application_df.head()
```

```python
# Determine the number of unique values in each column.
application_df.nunique()
```

```python
# Look at SPECIAL_CONSIDERATIONS value counts for binning
application_df[['SPECIAL_CONSIDERATIONS']].value_counts()
```

```python
# Drop the column, 'SPECIAL_CONSIDERATIONS'.
application_df = application_df.drop(columns=['SPECIAL_CONSIDERATIONS'])
```

```python
# Look at INCOME_AMT value counts for binning
application_df[['INCOME_AMT']].value_counts()
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `income_to_replace'
income_to_replace = ['25000-99999', '100000-499999', '1M-5M', '1-9999', '10000-24999', '10M-50M', '5M-10M', '50M+']

# Replace in dataframe
for money in income_to_replace:
    application_df['INCOME_AMT'] = application_df['INCOME_AMT'].replace(money,"With_Income")

# Check to make sure binning was successful
application_df['INCOME_AMT'].value_counts()
```

```python
# Look at APPLICATION_TYPE value counts for binning
application_df[['APPLICATION_TYPE']].value_counts()
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
application_types_to_replace = ['T17', 'T15', 'T29', 'T25',  'T14', 'T2', 'T12', 'T13', 'T9']

# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
application_df[['CLASSIFICATION']].value_counts()
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
application_df['CLASSIFICATION'].value_counts()[application_df['CLASSIFICATION'].value_counts() > 1]
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# The variable name `classifications_to_lessThan1000` are all classifications that have less than 1000 applications.
# The variable name `classifications_to_lessThan10k` are all classifications that have less than 10,000 applications.
classifications_to_lessThan1000 = ['C7000', 'C2800','C7100','C1300','C1280','C1230','C1400','C7200','C2300','C1240','C8000','C7120','C1500','C1800','C6000','C1250','C8200','C1238','C1278','C1235','C1237','C7210','C2400','C1720','C4100','C1257','C1600','C1260','C2710','C0','C3200','C1234','C1246','C1267','C1256', 'C1700', 'C4000', 'C5000', 'C1270', 'C2700', 'C1370',
'C1236', 'C2190', 'C4200', 'C5200', 'C1900', 'C2600', 'C1248', 'C6100', 'C1820', 'C1580', 'C2570', 'C1732', 'C3700', 'C2500', 'C1570', 'C1283', 'C2380', 'C1728', 'C2170', 'C4120', 'C8210', 'C4500', 'C1245', 'C2561', 'C2150']
classifications_to_lessThan10k = ['C2000' , 'C1200' , 'C3000' , 'C2100']

# Replace in dataframe
for cls in classifications_to_lessThan1000:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other, Less Than 100")

# Replace in dataframe
for cls in classifications_to_lessThan10k:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other, Less Than 10K")

# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
#Converting all the categorical data into numeric values
test_app_df = pd.get_dummies(application_df)
```

```python
# Split our preprocessed data into our features and target arrays
y = test_app_df.IS_SUCCESSFUL.values
X = test_app_df.drop(columns=['IS_SUCCESSFUL'])

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, stratify=y)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
num_input_feat = 31
hidden_nodes_layer1 =  80
hidden_nodes_layer2 = 50
hidden_nodes_layer3 = 30

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer1,  input_dim=num_input_feat, activation="relu"))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="sigmoid"))

# Third hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer3, activation="relu"))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))
```

```python
# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
%time
fit_model = nn.fit(X_train_scaled,y_train,epochs=100).on_epoch_begin(5)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
